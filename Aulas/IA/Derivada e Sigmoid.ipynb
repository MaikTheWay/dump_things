````python
import numpy as np

# FunГ§ГЈo sigmoid
def sigmoid(soma):
    return 1 / (1 + np.exp(-soma))

# Derivada da sigmoid
def sigmoidDerivada(sig):
    return sig * (1 - sig)

# Entradas (XOR)
entradas = np.array([[0,0],
                     [0,1],
                     [1,0],
                     [1,1]])

# SaГӯdas esperadas
saida = np.array([[0],[1],[1],[0]])

# Pesos iniciais
peso0 = np.array([[-0.424, -0.740, -0.961],
                  [0.358, -0.577, -0.469]])

peso1 = np.array([[-0.017], [-0.893], [-0.148]])

# HiperparГўmetros <3
epocas = 100
taxa_aprendizagem = 0.1

print("=== TREINAMENTO DA REDE ===\n")
for epoca in range(epocas):
    camadaEntrada = entradas
    somaSinapse0 = np.dot(camadaEntrada, peso0)
    camadaOculta = sigmoid(somaSinapse0)
    somaSinapse1 = np.dot(camadaOculta, peso1)
    camadaSaida = sigmoid(somaSinapse1)
    erroCamadaSaida = saida - camadaSaida
    mediaErro = np.mean(np.abs(erroCamadaSaida))
    derivadaSaida = sigmoidDerivada(camadaSaida)
    deltaSaida = erroCamadaSaida * derivadaSaida
    peso1Transposta = peso1.T
    deltaSaidaXPeso = deltaSaida.dot(peso1Transposta)
    deltaCamadaOculta = deltaSaidaXPeso * sigmoidDerivada(camadaOculta)

    # AtualizaГ§ГЈo dos pesos
    peso1 += camadaOculta.T.dot(deltaSaida) * taxa_aprendizagem
    peso0 += camadaEntrada.T.dot(deltaCamadaOculta) * taxa_aprendizagem

    # Print a cada 10 Г©pocas
    if epoca % 10 == 0:
        print(f"Гүpoca {epoca} - Erro mГ©dio: {mediaErro:.5f}")

print("\n=== RESULTADOS FINAIS ===")
print("SaГӯda da rede:")
print(camadaSaida)

print("\nErro final:")
print(erroCamadaSaida)

print("\n=== SITUAГҮГғO PROBLEMA ===")
# Dados do EXERCICIO!
entrada = np.array([0.5, 0.5])
target = 0
output = 0.45
deriv_sigmoide = 0.2475
eta = 0.01

# CГЎlculo do delta
erro = target - output
delta_saida = erro * deriv_sigmoide
delta_saida_com_eta = delta_saida * eta

print(f"Erro: {erro}")
print(f"Delta (sem learning rate): {delta_saida}")
print(f"Delta (com learning rate): {delta_saida_com_eta}")

# InterpretaГ§ГЈo simples
if delta_saida < 0:
    print("\nInterpretaГ§ГЈo: A saГӯda estГЎ maior que o esperado, entГЈo os pesos devem ser reduzidos.")
else:
    print("\nInterpretaГ§ГЈo: A saГӯda estГЎ menor que o esperado, entГЈo os pesos devem ser aumentados.")
````

Quando o Delta da camada escondida pode ser zero?

```
Isso pode acontecer quando o erro propagado da camada seguinte Г© zero, quando os pesos que vГӘm da prГіxima camada sГЈo zero, ou quando a derivada da sigmoide na camada escondida Г© zero por saturaГ§ГЈo.
Isso Г© um grande problema porque o neurГҙnio para de aprender, jГЎ que nГЈo recebe sinal Гәtil para atualizar seus pesos.
```





DiferenГ§a entre Delta da saГӯda e Delta da camada escondida:

```
Na camada de saГӯda, o delta Г© calculado diretamente com base no erro entre saГӯda e target.
Na camada escondida, o delta vem do erro propagado da camada seguinte.
Isso Г© necessГЎrio porque neurГҙnios escondidos nГЈo tГӘm alvo direto; eles aprendem indiretamente pelo erro que contribuem para gerar.
```



Efeitos de um learning rate muito alto, como 1.0
```
Positivos: pode acelerar bastante o inГӯcio do aprendizado.
Negativos: pode fazer a rede ultrapassar o mГӯnimo, oscilar, ficar instГЎvel ou atГ© divergir. Na prГЎtica, o treinamento pode piorar em vez de melhorar.
```



Como o momentum ajuda

```
O momentum guarda uma вҖңmemГіriaвҖқ dos passos anteriores do gradiente (Acumulando um "impulso" das atualizaГ§Гөes anteriores dos pesos).
Ele ajuda a rede a continuar avanГ§ando em uma direГ§ГЈo Гәtil, reduzindo oscilaГ§Гөes e facilitando a saГӯda de mГӯnimos locais rasos. Na atualizaГ§ГЈo dos pesos, ele adiciona uma parcela do passo anterior ao passo atual.
```

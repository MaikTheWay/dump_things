```python
import numpy as np
import math
```

```python
# Função de ativação (Sigmoid)
# -------------------------------
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Dados de entrada (XOR)
# -------------------------------
entradas = np.array([
    [0,0],
    [1,0],
    [0,1],
    [1,1]
])

saidas = np.array([
    [0],
    [1],
    [1],
    [0]
])

print("\nEntradas:")
print(entradas)
print("\nSaídas esperadas:")
print(saidas)

# Pesos da rede
# -------------------------------

# Camada de entrada → camada oculta (2 entradas x 3 neurônios)
peso0 = np.array([
    [-0.424, -0.740, 0.961],
    [0.350, -0.577, 0.469]
])

# Camada oculta → camada de saída (3 neurônios x 1 saída)
peso1 = np.array([
    [-0.017],
    [-0.893],
    [-0.148]
])

print("\nPesos entrada → camada oculta:")
print(peso0)

print("\nPesos camada oculta → saída:")
print(peso1)

# Treinamento
# -------------------------------
epocas = 100
for epoca in range(epocas):

    # Camada de entrada
    camadaEntrada = entradas
    # Soma das sinapses camada oculta
    somaSinapse0 = np.dot(camadaEntrada, peso0)
    # Ativação camada oculta
    camadaOculta = sigmoid(somaSinapse0)
    # Soma das sinapses camada de saída
    somaSinapse1 = np.dot(camadaOculta, peso1)
    # Ativação camada de saída
    camadaSaida = sigmoid(somaSinapse1)
    # Cálculo do erro
    erroCamadaSaida = saidas - camadaSaida

    # Média absoluta do erro
    mediaAbsoluta = np.mean(np.abs(erroCamadaSaida))
    print(f"Época {epoca} | Erro médio: {mediaAbsoluta:.6f}")


# Resultados finais
# -------------------------------

print("\n==============================")
print("RESULTADOS FINAIS DA REDE")
print("==============================")

print("\nSoma das sinapses - camada oculta:")
print(somaSinapse0)

print("\nAtivação da camada oculta (sigmoid):")
print(camadaOculta)

print("\nSoma das sinapses - camada de saída:")
print(somaSinapse1)

print("\nSaída da rede após sigmoid:")
print(camadaSaida)

print("\nErro da camada de saída:")
print(erroCamadaSaida)

print("\nMédia absoluta do erro final:")
print(mediaAbsoluta)
```

```python
#criando a função de sigmoid
def sigmoid(soma):
  return 1/(1+np.exp(-soma))
```

```python
#criando a derivada da função
def sigmoidDerivada(sig):
  return sig*(1-sig)
```

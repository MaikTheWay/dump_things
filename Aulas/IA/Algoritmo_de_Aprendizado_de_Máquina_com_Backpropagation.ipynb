```python
# import Libs
import numpy as np

def sigmoid(soma):
  return 1 / (1+np.exp(-soma))

def sigmoidDerivada(sig):
  return sig*(1-sig)

entradas = np.array([[0,0],
                    [0,1],
                    [1,0],
                    [1,1]])

# Valores desejados!
saidas = np.array([[0],[1],[1],[0]])

# Pesos(sinapses) para aprender os padrões da camada de entrada -> camada oculta
pesos0 = np.array([[-0.424, -0.740, -0.961],
                   [0.358, -0.577, -0.469]])
print(f'Pesos da camada de entrada para a camada oculta: {pesos0}')
print('\n')

# Pesos(sinapses) para aprender os padrões da camada oculta -> camada de saida
pesos1 = np.array([[-0.017, -0.893, -0.148]])
print(f'Pesos da camada oculta para a camada de saida: {pesos1}')
print('\n')

# Criar quantidade de vezes que pode fazer o looping
epocas = 100
taxaAprendizagem = 0.3
momento = 1

for i in range(epocas):
  camadaEntrada = entradas
  somaSinapse0 = np.dot(camadaEntrada, pesos0)
  camadaOculta = sigmoid(somaSinapse0)

  somaSinapse1 = np.dot(camadaOculta, pesos1.T)
  camadaSaida = sigmoid(somaSinapse1)

  erroCamadaSaida = saidas - camadaSaida
  mediaAbsoluta = np.mean(np.abs(erroCamadaSaida))
  print(f"Erro: {mediaAbsoluta}")

  derivadaSaida = sigmoidDerivada(camadaSaida)
  deltaSaida = erroCamadaSaida * derivadaSaida

  # Usar pesos1 (1,3) para backpropagar o erro para a camada oculta
  deltaSaidaXPeso = deltaSaida.dot(pesos1)

  # Usar deltaSaidaXPeso para calcular deltaCamadaOculta
  deltaCamadaOculta = deltaSaidaXPeso * sigmoidDerivada(camadaOculta)

  # Cálculo do pesoNovo1 para manter a forma de pesos1 (1,3)
  pesoNovo1 = deltaSaida.T.dot(camadaOculta)
  pesos1 = (pesos1 * momento) + (pesoNovo1 * taxaAprendizagem)

  # camdaEntradaTransposta deve ser camadaEntrada.T
  pesosNovo0 = camadaEntrada.T.dot(deltaCamadaOculta)
  pesos0 = (pesos0 * momento) + (pesosNovo0 * taxaAprendizagem)
```

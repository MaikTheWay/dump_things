```python
#Importar bibliotecas
import numpy as np
```

```python
#Parametros da rede neura Perceptron de uma camada
entradas = np.array([1, 7, 5])
pesos = np.array([0.8, 0.1, 0])
```

```python
#Cria uma função que faz a somatoria da rede neral Perceptron de uma camada
def soma(e, p): #e=entrada e p=peso
   return e.dot(p)
#dot product / produto escalar
```

```python
s = soma(entradas, pesos)
```

```python
print(f'O resultado da somatória do perceptron é {s}')
```

```python
# Função de ativação, na qual vamos usar a função degrau
def stepFunction(soma):
    if(soma >= 1):
        return 1
    else:
        return 0
```

```python
r = stepFunction(s)
```

```python
print(f'O resultado da ativação degrau é {r}')
```

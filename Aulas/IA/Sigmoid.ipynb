Exercicio 1

Imagine que agora chega um cliente que gosta de café Suave (0) e com leite (1). Sem fazer todos os cálculos novamente, você acha que a probabilidade final seria a mesma, maior ou menor? Por quê? Qual conceito importante sobre redes neurais isso demonstra?

```python
# Função sigmoide
def signoid(soma):
  return 1/(1+np.exp(-soma))

# ----- Exemplo Aula
#  a = signoid(0)
#  print(a)

# Entradas
x1 = 0  #suave
x2 = 1  #leite

# Pesos camada entrada -> oculta
w_x1_h1 = 1.0
w_x2_h1 = -1.0

w_x1_h2 = -1.0
w_x2_h2 = 1.0

# Pesos camada oculta -> saída
w_h1_y = 0.5
w_h2_y = 0.5

# ---- Camada oculta
s_h1 = (x1 * w_x1_h1) + (x2 * w_x2_h1)
h1 = sigmoid(s_h1)

s_h2 = (x1 * w_x1_h2) + (x2 * w_x2_h2)
h2 = sigmoid(s_h2)

# ---- Camada de saída
s_y = (h1 * w_h1_y) + (h2 * w_h2_y)
y = sigmoid(s_y)

print("h1:", round(h1, 3))
print("h2:", round(h2, 3))
print("Probabilidade para o pessoal gostar do café:", round(y, 3))

# Sigmoide
# Só gera valores positivos (0 até 1).
# Muito usada quando queremos probabilidade.
# Exemplo: 0.62 = 62% de chance.

# Sigmoide transforma valores em probabilidades (0 a 1).
```

Seria diferente (provavelmente menor).

Porque ao mudar as entradas para x1 = 0 (suave) e x2 = 1 (leite), mudamos a soma ponderada nos neurônios: s=(entrada)x(peso)

Como os pesos são diferentes para cada entrada, o valor que entra na função sigmoide também muda

Exercicio 2

Imagine que agora chega um cliente que gosta de café Suave (0) e com leite (1). Sem fazer todos os cálculos novamente, você acha que a probabilidade final seria a mesma, maior ou menor? Por quê? Qual conceito importante sobre redes neurais isso demonstra?

```python
# Função sigmoide
def signoid(soma):
  return 1/(1+np.exp(-soma))

# ----- Exemplo Aula
#  a = signoid(0)
#  print(a)

# Entradas
x1 = 0  #suave
x2 = 1  #leite

# Pesos camada entrada -> oculta
w_x1_h1 = 5.0
w_x2_h1 = -5.0

w_x1_h2 = -5.0
w_x2_h2 = 5.0

# Pesos camada oculta -> saída
w_h1_y = 0.5
w_h2_y = 0.5

# ---- Camada oculta
s_h1 = (x1 * w_x1_h1) + (x2 * w_x2_h1)
h1 = sigmoid(s_h1)

s_h2 = (x1 * w_x1_h2) + (x2 * w_x2_h2)
h2 = sigmoid(s_h2)

# ---- Camada de saída
s_y = (h1 * w_h1_y) + (h2 * w_h2_y)
y = sigmoid(s_y)

print("h1:", round(h1, 3))
print("h2:", round(h2, 3))
print("Probabilidade para o pessoal gostar do café:", round(y, 3))
```

Os valores são determinados por:
Entradas (x1, x2)
Pesos da rede
Soma ponderada
Função de ativação (sigmoid)

Mudanças em qualquer um desses fatores mudam a ativação.

O que aconteceria se os pesos fossem 5.0 e -5.0?
Os valores da soma ficariam muito maiores ou menores, fazendo a sigmoide ir para os extremos:

valores próximos de 0
ou próximos de 1

sso acontece porque a sigmoide satura quando recebe valores muito grandes ou muito negativos.

Exercicio 3


Compare o resultado que você obteve (ex: ~0.73) com a tabela da função sigmoide. Se ao invés da sigmoide, tivéssemos usado a função Tangente Hiperbólica (tanh), que retorna valores entre -1 e 1, o que aconteceria com o valor final da probabilidade? Conseguiríamos interpretá-lo diretamente como uma probabilidade?

```python
import numpy as np

# Função tangente hiperbólica
def tanh(soma):
    return np.tanh(soma)

# ----- Exemplo Aula
#  a = signoid(0)
#  print(a)

# Entradas
x1 = 0  # suave
x2 = 1  # leite

# Pesos camada entrada -> oculta
w_x1_h1 = 1.0
w_x2_h1 = -1.0

w_x1_h2 = -1.0
w_x2_h2 = 1.0

# Pesos camada oculta -> saída
w_h1_y = 0.5
w_h2_y = 0.5

# ---- Camada oculta
s_h1 = (x1 * w_x1_h1) + (x2 * w_x2_h1)
h1 = tanh(s_h1)

s_h2 = (x1 * w_x1_h2) + (x2 * w_x2_h2)
h2 = tanh(s_h2)

# ---- Camada de saída
s_y = (h1 * w_h1_y) + (h2 * w_h2_y)
y = tanh(s_y)

print("h1:", round(h1, 3))
print("h2:", round(h2, 3))
print("Saída da rede (tanh):", round(y, 3))

# tanh
# Gera valores negativos e positivos (-1 até 1).
# Mostra se a ativação é negativa, neutra ou positiva.
# Não representa probabilidade diretamente.

# tanh transforma valores em ativações centradas em zero (-1 a 1)
```

A tanh retorna valores entre -1 e 1, enquanto a sigmoide retorna 0 a 1.

Mas não poderíamos interpretar diretamente como probabilidade, porque probabilidades precisam estar entre 0 e 1.

Sigmoid → usada para probabilidade
tanh → usada para ativações centradas em zero
Escolha da função de ativação depende do objetivo da rede.

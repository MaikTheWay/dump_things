Prompt Inicial: Calculadora Simples (CLI)

"Escreva um programa Python de linha de comando que funcione como uma calculadora simples. Ele deve permitir ao usuário escolher uma operação (adição, subtração, multiplicação ou divisão), inserir dois números e exibir o resultado. Trate erros como divisão por zero e entradas inválidas."

```python
def calcular():
    print("Calculadora Simples")
    print("Escolha a operação:")
    print("1 - Adição")
    print("2 - Subtração")
    print("3 - Multiplicação")
    print("4 - Divisão")

    operacao = input("Digite o número da operação desejada: ")

    try:
        num1 = float(input("Digite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))

        if operacao == '1':
            resultado = num1 + num2
        elif operacao == '2':
            resultado = num1 - num2
        elif operacao == '3':
            resultado = num1 * num2
        elif operacao == '4':
            if num2 == 0:
                raise ZeroDivisionError
            resultado = num1 / num2
        else:
            print("Operação inválida.")
            return

        print(f"Resultado: {resultado}")
    except ValueError:
        print("Erro: entrada inválida.")
    except ZeroDivisionError:
        print("Erro: divisão por zero.")

if __name__ == "__main__":
    calcular()
```

Prompt para Criar uma Interface Gráfica (GUI)

"Adapte a calculadora científica para usar uma interface gráfica no navegador utilizando a biblioteca Gradio, em vez de Tkinter. A calculadora deve permitir ao usuário escolher operações básicas (adição, subtração, multiplicação, divisão) e científicas (exponenciação, raiz quadrada, logaritmo). Use campos de entrada para os dois números e um menu suspenso para selecionar a operação. Mostre o resultado de forma clara. Trate erros como divisão por zero ou entradas inválidas."

```python
!pip install gradio --quiet

import gradio as gr
import math

def calcular(a, b, op):
    try:
        a = float(a)
        b = float(b)
        if op == '+':
            return a + b
        elif op == '-':
            return a - b
        elif op == '*':
            return a * b
        elif op == '/':
            return a / b if b != 0 else "Erro: divisão por zero"
        elif op == '**':
            return a ** b
        elif op == '√':
            return math.sqrt(a)
        elif op == 'log10':
            return math.log10(a)
    except Exception as e:
        return f"Erro: {e}"

iface = gr.Interface(
    fn=calcular,
    inputs=[
        gr.Number(label="Número 1"),
        gr.Number(label="Número 2"),
        gr.Dropdown(['+', '-', '*', '/', '**', '√', 'log10'], label="Operação")
    ],
    outputs="text",
    title="Calculadora com Gradio"
)

iface.launch()
```

Prompt para Escrever Números por Extenso

"Adicione uma nova funcionalidade à calculadora feita com Gradio: além de exibir o resultado numérico, a interface também deve mostrar esse número por extenso (em português). Por exemplo, se o resultado for 123, deve aparecer 'cento e vinte e três'. Utilize a biblioteca num2words com o idioma português brasileiro (pt_BR). O resultado por extenso deve ser exibido junto ao valor numérico em uma área de saída clara. Trate possíveis erros de conversão."

```python
!pip install gradio num2words --quiet

import gradio as gr
import math
from num2words import num2words

def calcular(a, b, op):
    try:
        a = float(a)
        b = float(b)

        # Cálculo
        if op == '+':
            resultado = a + b
        elif op == '-':
            resultado = a - b
        elif op == '*':
            resultado = a * b
        elif op == '/':
            if b == 0:
                return "Erro: divisão por zero", ""
            resultado = a / b
        elif op == '**':
            resultado = a ** b
        elif op == '√':
            resultado = math.sqrt(a)
        elif op == 'log10':
            resultado = math.log10(a)
        else:
            return "Operação inválida", ""

        # Número por extenso
        try:
            extenso = num2words(resultado, lang='pt_BR')
        except:
            extenso = "Não foi possível converter para extenso."

        return f"Resultado: {resultado}", f"Por extenso: {extenso}"

    except Exception as e:
        return f"Erro: {e}", ""

iface = gr.Interface(
    fn=calcular,
    inputs=[
        gr.Number(label="Número 1"),
        gr.Number(label="Número 2"),
        gr.Dropdown(['+', '-', '*', '/', '**', '√', 'log10'], label="Operação")
    ],
    outputs=[
        gr.Textbox(label="Resultado Numérico"),
        gr.Textbox(label="Resultado por Extenso")
    ],
    title="Calculadora com Gradio e Número por Extenso",
    description="Escolha uma operação e veja o resultado tanto numérico quanto por extenso (em português)."
)

iface.launch()
```

Gravar nome do recordista e tempo que ele demorou

"Adicione ao jogo da forca um sistema de recordes: ao final de uma vitória, o programa deve perguntar o nome do jogador e registrar o tempo que ele levou para vencer a partida. Salve esses dados em um arquivo local (ex: 'recordes.txt'). Se o tempo for menor que o recorde atual, informe que ele quebrou o recorde e mostre o recordista anterior."

```python
import random
import time

def desenhar_forca(erros):
    estagios = [
        """
           ------
           |    |
           |
           |
           |
           |
        --------
        """,
        """
           ------
           |    |
           |    O
           |
           |
           |
        --------
        """,
        """
           ------
           |    |
           |    O
           |    |
           |
           |
        --------
        """,
        """
           ------
           |    |
           |    O
           |   /|
           |
           |
        --------
        """,
        """
           ------
           |    |
           |    O
           |   /|\\
           |
           |
        --------
        """,
        """
           ------
           |    |
           |    O
           |   /|\\
           |   /
           |
        --------
        """,
        """
           ------
           |    |
           |    O
           |   /|\\
           |   / \\
           |
        --------
        """
    ]
    print(estagios[erros])

def gravar_recorde(nome, tempo):
    try:
        with open("recordes.txt", "r") as arquivo:
            linhas = arquivo.readlines()
            recordes = [linha.strip().split(",") for linha in linhas]
            # Convertendo tempos para float
            recordes = [(r[0], float(r[1])) for r in recordes]
    except FileNotFoundError:
        recordes = []

    # Checar se o tempo atual é recorde
    tempos = [r[1] for r in recordes]
    if not tempos or tempo < max(tempos) or len(recordes) < 5:
        recordes.append((nome, tempo))
        recordes = sorted(recordes, key=lambda x: x[1])[:5]  # Mantém top 5 recordes
        with open("recordes.txt", "w") as arquivo:
            for r in recordes:
                arquivo.write(f"{r[0]},{r[1]:.2f}\n")
        print(f"\n🎉 Novo recorde! Parabéns {nome}, com tempo de {tempo:.2f} segundos.\n")
        print("🏆 Top 5 recordes:")
        for i, (n, t) in enumerate(recordes, 1):
            print(f"{i}. {n} - {t:.2f} segundos")
    else:
        print(f"\nSeu tempo foi {tempo:.2f} segundos.")
        print("🏆 Top 5 recordes:")
        for i, (n, t) in enumerate(recordes, 1):
            print(f"{i}. {n} - {t:.2f} segundos")

def escolher_palavra(nivel):
    # Lista de palavras (exemplo; em prática pode ser maior)
    palavras = [
        "amor", "paz", "sol", "lua", "mar", "flor", "rio", "céu", "mel", "voz",
        "tempo", "livro", "carro", "casa", "gato", "cão", "noite", "dia", "luz", "vento",
        "praia", "montanha", "floresta", "cidade", "amigo", "família", "escola", "trabalho",
        "festa", "música", "dança", "arte", "vida", "sonho", "coração", "alegria", "dor",
        "sorriso", "chave", "porta", "janela", "cadeira", "mesa", "computador", "telefone",
        "janela", "rua", "ponte", "praça", "tempo", "livro", "pássaro", "cachorro",
        "gato", "leão", "tigre", "elefante", "girafa", "cavalo", "vaca", "ovelha", "coelho"
    ]

    if nivel == "fácil":
        filtradas = [p for p in palavras if len(p) <= 4]
    elif nivel == "médio":
        filtradas = [p for p in palavras if 5 <= len(p) <= 7]
    elif nivel == "difícil":
        filtradas = [p for p in palavras if len(p) > 7]
    else:
        filtradas = palavras  # sem filtro

    if not filtradas:
        filtradas = palavras  # fallback se não achar nenhuma palavra

    return random.choice(filtradas)

def jogo_forca():
    print("=== Bem-vindo ao Jogo da Forca ===")
    print("Escolha o nível de dificuldade: fácil, médio ou difícil")
    nivel = input("Nível: ").strip().lower()

    palavra = escolher_palavra(nivel)
    letras_adivinhadas = set()
    tentativas = 6
    erros = 0

    print(f"\nA palavra tem {len(palavra)} letras.")

    inicio = time.time()

    while erros < tentativas:
        desenhar_forca(erros)
        palavra_mostrada = " ".join([letra if letra in letras_adivinhadas else "_" for letra in palavra])
        print("Palavra:", palavra_mostrada)

        chute = input("Digite uma letra: ").lower()

        if len(chute) != 1 or not chute.isalpha():
            print("Digite apenas uma letra válida.")
            continue

        if chute in letras_adivinhadas:
            print("Você já tentou essa letra.")
            continue

        if chute in palavra:
            letras_adivinhadas.add(chute)
            print("Acertou!")
        else:
            erros += 1
            print(f"Errou! Tentativas restantes: {tentativas - erros}")

        if all(letra in letras_adivinhadas for letra in palavra):
            fim = time.time()
            duracao = fim - inicio
            print(f"\n🎉 Parabéns! Você acertou a palavra '{palavra}'!")
            nome = input("Digite seu nome para registrar o recorde: ")
            gravar_recorde(nome, duracao)
            break
    else:
        desenhar_forca(erros)
        print(f"\nFim de jogo! A palavra era '{palavra}'.")

if __name__ == "__main__":
    jogo_forca()
```

#PROMPT:#
Crie um programa em Python 3.10 ou superior chamado conversor_unidades.py. O programa será um conversor de unidades interativo via linha de comando (CLI).

O programa deve ser modular, escalável e robusto, permitindo que o usuário converta entre diferentes sistemas de medidas de forma simples e confiável.

O programa deve usar apenas bibliotecas nativas do Python, sem dependências externas.

A arquitetura do código deve ser composta por:

Funções específicas para cada tipo de conversão.

Um dicionário central `conversoes` que associa opções numéricas (inteiros) a tuplas contendo a descrição da conversão (string) e a função Python correspondente.

Uma função `menu()` responsável por exibir dinamicamente as opções disponíveis, listando-as a partir do dicionário `conversoes` em ordem crescente de chaves, seguida da opção "0. Sair".

Um loop principal `while True` que controla o fluxo de execução. Este loop deve:
1. Chamar a função `menu()` para exibir as opções.
2. Solicitar ao usuário que escolha uma opção, capturando a entrada como um número inteiro.
3. Tratar a entrada do usuário:
    - Se a opção for `0`, exibir "Saindo..." e encerrar o loop (`break`).
    - Se a opção não for encontrada nas chaves do dicionário `conversoes`, exibir "Opção inválida!".
    - Utilizar um bloco `try-except ValueError` para capturar entradas não numéricas para a opção do menu, exibindo "Entrada inválida, digite um número!".
    - Em caso de entrada inválida (opção não existente ou não numérica), o programa deve retornar ao início do loop, exibindo o menu novamente (`continue`).
4. Se a opção for válida (existir no dicionário `conversoes`):
    - Solicitar ao usuário que digite o valor a ser convertido, capturando a entrada como um número decimal (float).
    - Utilizar um bloco `try-except ValueError` para capturar entradas não numéricas para o valor a ser convertido, exibindo "Entrada inválida, digite um número!".

    - Chamar a função de conversão com o valor fornecido pelo usuário.
    - Exibir o resultado no formato exato: "Descrição: valor de entrada -> valor convertido", substituindo "Descrição", "valor de entrada" e "valor convertido" pelos valores correspondentes.

Todas as mensagens e textos exibidos ao usuário (menu, prompts, mensagens de erro, resultados) devem estar estritamente em português do Brasil.

O programa deve suportar as seguintes categorias de conversão, com as respectivas funções e entradas no dicionário `conversoes`:

1.  **Comprimento / Distância**
    - Metros para quilômetros (`metros_para_km`)
    - Quilômetros para metros (`km_para_metros`)
    - Quilômetros para milhas (`km_para_milhas`)

    - Milhas náuticas para quilômetros (`milha_nautica_para_km`)
    - Quilômetros para milhas náuticas (`km_para_milha_nautica`)
    - Pés para quilômetros (`pes_para_km`)
    - Quilômetros para pés (`km_para_pes`)
    - Jardas para quilômetros (`jardas_para_km`)

2.  **Temperatura**
    - Celsius para Fahrenheit (`celsius_para_fahrenheit`)
    - Fahrenheit para Celsius (`fahrenheit_para_celsius`)

3.  **Massa / Peso**
    - Gramas para quilogramas (`gramas_para_kg`)

4.  **Área**
    - Hectares para metros quadrados (`hectare_para_m2`)

O resultado final deve ser apenas o código-fonte completo em Python 3.10+ do arquivo `conversor_unidades.py`, sem quaisquer comentários adicionais no código (exceto docstrings se desejar para as funções de conversão, embora o prompt original as omita), explicações externas ou saídas que não sejam do próprio programa interativo. O código deve estar pronto para execução direta no terminal.

```python
# --- Funções de Conversão ---
def metros_para_km(m): return m / 1000
def km_para_metros(km): return km * 1000

def km_para_milhas(km): return km * 0.621371
def milhas_para_km(mi): return mi / 0.621371

def celsius_para_fahrenheit(c): return (c * 9/5) + 32
def fahrenheit_para_celsius(f): return (f - 32) * 5/9
def celsius_para_kelvin(c): return c + 273.15
def kelvin_para_celsius(k): return k - 273.15

def gramas_para_kg(g): return g / 1000
def kg_para_gramas(kg): return kg * 1000
def kg_para_libras(kg): return kg * 2.20462
def libras_para_kg(lb): return lb / 2.20462

def hectare_para_m2(ha): return ha * 10000
def m2_para_hectare(m2): return m2 / 10000
def acre_para_m2(ac): return ac * 4046.86
def m2_para_acre(m2): return m2 / 4046.86

def milha_nautica_para_km(mn): return mn * 1.852
def km_para_milha_nautica(km): return km / 1.852
def pes_para_km(ft): return ft * 0.0003048
def km_para_pes(km): return km / 0.0003048
def jardas_para_km(y): return y * 0.0009144
def km_para_jardas(km): return km / 0.0009144

# --- Dicionário de Opções ---
conversoes = {
    1: ("Metros -> Km", metros_para_km),
    2: ("Km -> Metros", km_para_metros),
    3: ("Km -> Milhas", km_para_milhas),
    4: ("Milhas -> Km", milhas_para_km),
    5: ("Celsius -> Fahrenheit", celsius_para_fahrenheit),
    6: ("Fahrenheit -> Celsius", fahrenheit_para_celsius),
    7: ("Celsius -> Kelvin", celsius_para_kelvin),
    8: ("Kelvin -> Celsius", kelvin_para_celsius),
    9: ("Gramas -> Kg", gramas_para_kg),
    10: ("Kg -> Gramas", kg_para_gramas),
    11: ("Kg -> Libras", kg_para_libras),
    12: ("Libras -> Kg", libras_para_kg),
    13: ("Hectare -> m²", hectare_para_m2),
    14: ("m² -> Hectare", m2_para_hectare),
    15: ("Acre -> m²", acre_para_m2),
    16: ("m² -> Acre", m2_para_acre),
    17: ("Milha náutica -> Km", milha_nautica_para_km),
    18: ("Km -> Milha náutica", km_para_milha_nautica),
    19: ("Pés -> Km", pes_para_km),
    20: ("Km -> Pés", km_para_pes),
    21: ("Jardas -> Km", jardas_para_km),
    22: ("Km -> Jardas", km_para_jardas)
}

# --- Menu ---
def menu():
    print("\n=== Conversor de Unidades ===")
    for opcao, (descricao, _) in conversoes.items():
        print(f"{opcao}. {descricao}")
    print("0. Sair")

# --- Loop Principal ---
while True:
    menu()
    try:
        opcao = int(input("Escolha uma opção: "))
        if opcao == 0:
            print("Saindo...")
            break
        if opcao not in conversoes:
            print("Opção inválida!")
            continue

        valor = float(input("Digite o valor: "))
        descricao, func = conversoes[opcao]
        print(f"{descricao}: {valor} -> {func(valor)}")
    except ValueError:
        print("Entrada inválida, digite um número!")
```

O script deve estar pronto para ser executado diretamente via `python gerador_senhas.py` no terminal.

Incluir a estrutura básica com funções para `gerar_senha`, `criptografar_senha`, `descriptografar_senha`, uma função auxiliar `perguntar_sim_nao`, o conjunto global `senhas_geradas`, e um bloco `if __name__ == "__main__":` chamando uma função principal (`main()`) que orquestra a interação e os passos de geração, criptografia e exibição.


```python
import secrets
import string
from cryptography.fernet import Fernet

# -------------------------------
# Armazena senhas únicas (em memória)
senhas_geradas = set()

# -------------------------------
# Geração de chave de criptografia (deve ser salva se for descriptografar depois)
chave = Fernet.generate_key()
fernet = Fernet(chave)

# -------------------------------
def gerar_senha(comprimento=12, usar_maiusculas=True, usar_minusculas=True,
                usar_numeros=True, usar_simbolos=True):

    if not (usar_maiusculas or usar_minusculas or usar_numeros or usar_simbolos):
        raise ValueError("Selecione pelo menos um tipo de caractere.")

    alfabeto = ''
    if usar_maiusculas:
        alfabeto += string.ascii_uppercase
    if usar_minusculas:
        alfabeto += string.ascii_lowercase
    if usar_numeros:
        alfabeto += string.digits
    if usar_simbolos:
        alfabeto += string.punctuation

    while True:
        senha = ''.join(secrets.choice(alfabeto) for _ in range(comprimento))
        if senha not in senhas_geradas:
            senhas_geradas.add(senha)
            return senha

# -------------------------------
def criptografar_senha(senha):
    senha_bytes = senha.encode()
    senha_criptografada = fernet.encrypt(senha_bytes)
    return senha_criptografada

# -------------------------------
def descriptografar_senha(senha_criptografada):
    senha_original = fernet.decrypt(senha_criptografada)
    return senha_original.decode()

# -------------------------------
def perguntar_sim_nao(texto):
    while True:
        resposta = input(texto + " (s/n): ").strip().lower()
        if resposta in ['s', 'sim', "ss"]:
            return True
        elif resposta in ['n', 'nao', 'não']:
            return False
        else:
            print("Digite apenas 's' ou 'n'.")

# -------------------------------
# INTERFACE VIA TERMINAL
def main():
    print("🔐 Gerador de Senhas Seguras 🔐\n")

    while True:
        try:
            comprimento = int(input("Informe o comprimento da senha (mínimo 4): "))
            if comprimento < 4:
                print("Comprimento muito curto. Tente novamente.")
                continue
            break
        except ValueError:
            print("Por favor, digite um número inteiro.")

    usar_maiusculas = perguntar_sim_nao("Incluir letras maiúsculas?")
    usar_minusculas = perguntar_sim_nao("Incluir letras minúsculas?")
    usar_numeros = perguntar_sim_nao("Incluir números?")
    usar_simbolos = perguntar_sim_nao("Incluir símbolos?")

    try:
        senha = gerar_senha(comprimento, usar_maiusculas, usar_minusculas, usar_numeros, usar_simbolos)
    except ValueError as e:
        print(e)
        return

    senha_criptografada = criptografar_senha(senha)
    senha_descriptografada = descriptografar_senha(senha_criptografada)

    print("\n🧾 RESULTADO:")
    print("Senha gerada:", senha)
    print("Senha criptografada (Fernet):", senha_criptografada.decode())
    print("Senha descriptografada:", senha_descriptografada)
    print("Chave de criptografia (salve isso se quiser descriptografar depois):", chave.decode())

# -------------------------------
if __name__ == "__main__":
    main()
```

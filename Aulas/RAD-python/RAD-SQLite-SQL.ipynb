# **Tutorial Python para bancos de dados SQLite**

Tutorial de programas Python para trabalhar com uma base de dados SQLite usando consultas SQL simples e complexas, bem como operações de criação e remoção de tabelas.

Este tutorial cobre uma ampla gama de operações básicas e avançadas para manipular dados usando Python e SQLite, oferecendo uma boa base para você começar a trabalhar com bancos de dados em seus projetos Python.

## **Criar e Conectar ao Banco de Dados**

Este programa conecta a um banco de dados SQLite, criando-o se não existir. O cursor permite executar comandos SQL. A conexão é fechada após a operação.

```python
import sqlite3

# Conectar ao banco de dados (ou criar se não existir)
conn = sqlite3.connect('meu_banco.db')

# Criar um cursor
cursor = conn.cursor()

print("Banco de dados conectado com sucesso!")

# Fechar a conexão
conn.close()
```

## **Criar uma Tabela Simples**

Este programa cria a tabela usuarios com colunas id, nome, e idade, se ainda não existir. O commit salva as alterações no banco de dados.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Criar uma tabela
cursor.execute('''
CREATE TABLE IF NOT EXISTS usuarios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    idade INTEGER NOT NULL
)
''')

print("Tabela criada com sucesso!")

res = conn.commit()
conn.close()
```

## **Inserir Dados em uma Tabela**

Este programa insere três registros na tabela usuarios. Os dados são salvos no banco de dados com commit e a conexão é fechada.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Inserir dados
cursor.execute('''
INSERT INTO usuarios (nome, idade) VALUES
('Alice', 30),
('Bob', 25),
('Carol', 40)
''')

print("Dados inseridos com sucesso!")

conn.commit()
conn.close()
```

## **Selecionar Todos os Dados de uma Tabela**

Este programa consulta e imprime todos os registros da tabela usuarios. A função fetchall recupera todos os resultados da consulta.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Selecionar todos os dados
cursor.execute('SELECT * FROM usuarios')
resultados = cursor.fetchall()

for row in resultados:
    print(row)

conn.close()
```

## **Selecionar Dados com Condição**

Aqui, apenas os registros em que idade é maior que 30 são selecionados e exibidos. O uso de WHERE permite aplicar essa condição.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Selecionar dados com condição
cursor.execute('SELECT * FROM usuarios WHERE idade > 30')
resultados = cursor.fetchall()

for row in resultados:
    print(row)

conn.close()
```

## **Atualizar Dados em uma Tabela**

Este programa atualiza o campo idade do usuário com nome = "Bob" para 35. As alterações são salvas com commit.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Atualizar dados
cursor.execute('UPDATE usuarios SET idade = 35 WHERE nome = "Bob"')

print("Dados atualizados com sucesso!")

conn.commit()
conn.close()
```

## **Deletar Dados de uma Tabela**

Este programa exclui o registro onde nome = "Carol". A função DELETE remove o registro do banco de dados.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Deletar dados
cursor.execute('DELETE FROM usuarios WHERE nome = "Carol"')

print("Dados deletados com sucesso!")

conn.commit()
conn.close()
```

## **Criar Tabela com Relacionamento (Chave Estrangeira)**

Este programa cria a tabela pedidos, relacionando-a à tabela usuarios por meio de usuario_id, que é uma chave estrangeira.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Criar tabela com chave estrangeira
cursor.execute('''
CREATE TABLE IF NOT EXISTS pedidos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    descricao TEXT NOT NULL,
    usuario_id INTEGER,
    FOREIGN KEY(usuario_id) REFERENCES usuarios(id)
)
''')

print("Tabela com relacionamento criada com sucesso!")

conn.commit()
conn.close()
```

## **Inserir Dados em Tabela com Relacionamento**

Este programa insere registros na tabela pedidos, associando-os aos usuários correspondentes por meio de usuario_id.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Inserir dados em tabela relacionada
cursor.execute('''
INSERT INTO pedidos (descricao, usuario_id) VALUES
('Pedido 1', 1),
('Pedido 2', 2)
''')

print("Dados inseridos na tabela com relacionamento!")

conn.commit()
conn.close()
```

## **Consultar Dados com Junção (JOIN)**

 Este programa utiliza JOIN para combinar dados das tabelas usuarios e pedidos, exibindo os nomes dos usuários e suas descrições de pedidos.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Consultar dados com JOIN
cursor.execute('''
SELECT usuarios.nome, pedidos.descricao
FROM usuarios
JOIN pedidos ON usuarios.id = pedidos.usuario_id
''')
resultados = cursor.fetchall()

for row in resultados:
    print(row)

conn.close()
```

## **Usar Subconsultas**

Este programa utiliza uma subconsulta para selecionar os nomes de usuários que têm pedidos associados. A subconsulta é usada dentro do WHERE.


```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Usar subconsulta
cursor.execute('''
SELECT id, nome FROM usuarios WHERE id IN (SELECT usuario_id FROM pedidos)
''')
resultados = cursor.fetchall()

for row in resultados:
    print(row)

conn.close()
```

## **Criar Índices para Acelerar Consultas**

Este programa cria um índice na coluna nome da tabela usuarios, melhorando a velocidade das consultas que utilizam essa coluna.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Criar índice
cursor.execute('CREATE INDEX idx_usuario_nome ON usuarios (nome)')

print("Índice criado com sucesso!")

conn.commit()
conn.close()
```

## **Consultar Dados com Ordenação (ORDER BY)**

Este programa consulta a tabela usuarios e ordena os resultados pela coluna idade em ordem decrescente, exibindo-os na tela.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Consultar dados com ordenação
cursor.execute('SELECT * FROM usuarios ORDER BY idade DESC')
resultados = cursor.fetchall()

for row in resultados:
    print(row)

conn.close()
```

## **Consultar Dados com Limite (LIMIT)**

Este programa seleciona os dois primeiros registros da tabela usuarios usando a cláusula LIMIT, limitando a quantidade de resultados retornados.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Consultar dados com limite
cursor.execute('SELECT * FROM usuarios LIMIT 2')
resultados = cursor.fetchall()

for row in resultados:
    print(row)

conn.close()
```

## **Contar Registros em uma Tabela**

Este programa conta o número total de registros na tabela usuarios usando a função agregada COUNT.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Contar registros
cursor.execute('SELECT COUNT(*) FROM usuarios')
total = cursor.fetchone()[0]

print(f'Total de usuários: {total}')

conn.close()
```

## **Remover Tabela do Banco de Dados**

Este programa remove a tabela pedidos do banco de dados se ela existir, utilizando o comando DROP TABLE.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Remover tabela
cursor.execute('DROP TABLE IF EXISTS pedidos')

print("Tabela removida com sucesso!")

conn.commit()
conn.close()
```

## **Atualizar Estrutura de uma Tabela (ALTER TABLE)**

Este programa adiciona uma nova coluna email à tabela usuarios, modificando a estrutura da tabela existente.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Adicionar uma nova coluna
cursor.execute('ALTER TABLE usuarios ADD COLUMN email TEXT')

print("Tabela atualizada com sucesso!")

conn.commit()
conn.close()
```

## **Realizar Consultas Complexas com Funções Agregadas**

Este programa calcula a média de idade dos usuários na tabela usuarios utilizando a função agregada AVG.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

# Consulta com função agregada
cursor.execute('SELECT AVG(idade) FROM usuarios')
media_idade = cursor.fetchone()[0]

print(f'Média de idade: {media_idade}')

conn.close()
```

## **Transações: Executar Várias Consultas de Forma Atômica**

Este programa executa múltiplas operações de forma atômica usando transações. Se ocorrer um erro, as mudanças são desfeitas com rollback.

```python
import sqlite3

conn = sqlite3.connect('meu_banco.db')
cursor = conn.cursor()

try:
    # Iniciar transação
    conn.execute('BEGIN')

    cursor.execute('INSERT INTO usuarios (nome, idade) VALUES ("Dave", 29)')
    cursor.execute('INSERT INTO usuarios (nome, idade) VALUES ("Eve", 34)')

    # Confirmar transação
    conn.commit()
    print("Transação concluída com sucesso!")

except Exception as e:
    conn.rollback()
    print(f"Erro durante a transação: {e}")

finally:
    conn.close()
```

## **Conectar-se ao Banco de Dados em Memória (Para Testes)**

Este programa cria um banco de dados temporário em memória (RAM) para testes rápidos, permitindo a criação de tabelas e inserção de dados sem afetar um banco de dados real.

```python
import sqlite3

# Conectar ao banco de dados em memória
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Criar tabela para testes
cursor.execute('''
CREATE TABLE IF NOT EXISTS produtos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    preco REAL NOT NULL
)
''')

# Inserir dados de teste
cursor.execute('''
INSERT INTO produtos (nome, preco) VALUES
('Produto 1', 9.99),
('Produto 2', 19.99)
''')

cursor.execute('SELECT * FROM produtos')
resultados = cursor.fetchall()

for row in resultados:
    print(row)

conn.close()
```
